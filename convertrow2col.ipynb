```python
import csv
f = open("drive/MyDrive/Colab Notebooks/demofile.txt", "w")
with open('drive/MyDrive/Colab Notebooks/temporary.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print('By pass header')
            print(f'{row[0]}')
            print(f'{row}')
        else:
            # print(f'{row[0]}')
            # print(f'{row}')
            # print(f'{row[0]}\n')
            print(f'{row[1]}')
            f.write(f'{row[1]}\n')
            print(f'{row[2]}')
            f.write(f'{row[2]}\n')
            print(f'{row[3]}')
            f.write(f'{row[3]}\n')
            print(f'{row[4]}')
            f.write(f'{row[4]}\n')
            print(f'{row[5]}')
            f.write(f'{row[5]}\n')
            print(f'{row[6]}\n\n')
            f.write(f'{row[1]}\n\n')
        line_count += 1
    print(f'Processed {line_count} lines.')
f.close()
```
